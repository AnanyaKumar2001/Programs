#IMAGE TO PENCIL SKETCH WITH PYTHON

**Description  :**
      We need to read the image in RGB format and then convert it to a grayscale image. This will turn an image into a classic black and white photo. Then the next thing to do is invert the grayscale image also called negative image, this will be our inverted grayscale image. Inversion can be used to enhance details. Then we can finally create the pencil sketch by mixing the grayscale image with the blurry image. This can be done by dividing the grayscale image by the inverted blurry image.

###Importing Libraries

```python
import cv2
from google.colab.patches import cv2_imshow
```

### Input Image

```python
image = cv2.imread('/content/input.jpg')
```

### Printing input image

```python
print(cv2_imshow(image))
```

### Converting Original image to Grayscale image

```python
greyImage = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imwrite('grey_input.jpg', greyImage)
```

### Printing Grayscale image

```python
print(cv2_imshow(greyImage))
```

### Inverting the image

```python
inverted = 255 - greyImage
```

### Printing Inverted image

```python
print(cv2_imshow(inverted))
```

### Blur the image by using Gaussian function

```python
blur = cv2.GaussianBlur(inverted, (21, 21), 0)
```

### Inverting the blurred image

```python
invertBlur = 255 - blur
```

### Pencil Sketch

```python
sketch = cv2.divide(greyImage, invertBlur, scale=256)
cv2.imwrite('sketch.jpg', sketch)
```

### Printing Pencil sketch image

```python
print(cv2_imshow(sketch))
```
