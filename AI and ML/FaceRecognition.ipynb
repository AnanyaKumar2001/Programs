```python
ret,thresh_binary = cv2.threshold(img2, 127,255,cv2.THRESH_BINARY)
ret,thresh_binary_inv = cv2.threshold(img2, 127,255,cv2.THRESH_BINARY_INV)
ret,thresh_trunc = cv2.threshold(img2, 127,255,cv2.THRESH_TRUNC)
ret,thresh_tozero = cv2.threshold(img2, 127,255,cv2.THRESH_TOZERO)
ret,thresh_tozero_inv = cv2.threshold(img2, 127,255,cv2.THRESH_TOZERO_INV)
```

```python
names = ['ORIGINAL' , 'BINARY', 'THRESH_BINARY' ,'THRESH_TRUNC' , 'THRESH_TOZERO' , 'THRESH_TOZERO_INV' ]

images = img2,thresh_binary,thresh_binary_inv,thresh_trunc,thresh_tozero,thresh_tozero_inv
for i in range(6):
  plt.subplot(2,3,i+1), plt.imshow(images[i],'gray')
  plt.title(names[i])
  plt.xticks([]), plt.yticks([])
```

```python
face_cascade = cv2.CascadeClassifier('/content/haarcascade_frontalface_default.xml')
```

```python
faces = face_cascade.detectMultiScale(img3, 1.1, 4)

for (x, y, w, h) in faces:
  cv2.rectangle(img, (x,y), (x+w, y+h), (255, 0, 0), 2)

cv2_imshow(img)
cv2.waitKey()
```

## Multiple Face Recognition with different types of expressions

```python
img5 = cv2.imread('/content/11322327-young-woman-with-multiple-face-expressions.jpg')
print(cv2_imshow(img5))
```

```python
img6 = cv2.cvtColor(img5, cv2.COLOR_BGR2GRAY)
print(cv2_imshow(img6))
```

```python
faces = face_cascade.detectMultiScale(img6, 1.1, 4)

for (x, y, w, h) in faces:
  cv2.rectangle(img5, (x,y), (x+w, y+h), (255, 0, 0), 2)

cv2_imshow(img5)
cv2.waitKey()
```
